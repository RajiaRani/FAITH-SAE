Here we load the matched-budget config and compute the analytic **Causal Faithfulness Score (CFS)** across steering variants — the headline on-manifold-vs-naive saving, visible fully offline.

```
from src.utils import load_config
cfg = load_config('configs/default.yaml')
print(cfg)
```

```
# Analytic CFS comparison across steering methods (shared cost model in src.utils).
from src.utils import faithfulness
variants = ['onmanifold_steer', 'supervised_steer', 'naive_steer', 'clamp_steer', 'random_steer']
rows = {v: faithfulness(v, cfg) for v in variants}
print(f"{'variant':18s} {'mono':>6s} {'spec':>6s} {'suff':>6s} {'offres':>7s} {'CFS':>6s}")
for v in variants:
    r = rows[v]
    print(f"{v:18s} {r['monotonicity']:6.2f} {r['specificity']:6.2f} "
          f"{r['sufficiency']:6.2f} {r['offmanifold_residual']:7.2f} {r['cfs']:6.2f}")
print('\nHeadline: on-manifold CFS >> naive/clamp/random (supervised = quality reference).')
```

```
# Bar chart of analytic CFS by steering variant (placeholder until real M3 runs).
import matplotlib.pyplot as plt
vals = [rows[v]['cfs'] for v in variants]
plt.figure(figsize=(6,3))
plt.bar(variants, vals)
plt.ylabel('CFS  (harmonic mean, [0,1])'); plt.ylim(0,1)
plt.xticks(rotation=30, ha='right'); plt.title('Analytic CFS by steering method (illustrative)')
plt.tight_layout(); plt.show()
```
